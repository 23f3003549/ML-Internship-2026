## Total trips

```python
len(df)
df.shape[0]
```

```python
df['pickup_datetime']=pd.to_datetime(df["pickup_datetime"])

df['hour'] = df["pickup_datetime"].dt.hour
df['day'] = df['pickup_datetime'].dt.day
df['weekday'] = df['pickup_datetime'].dt.day_name()
df['month'] = df['pickup_datetime'].dt.month
df['year'] = df['pickup_datetime'].dt.year

weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
```

# TRIP DISTRIBUTION BY HOUR

```python
trips_per_hour = df.groupby('hour').size().sort_index()
trips_per_hour
```

```python
max_hour=trips_per_hour.idxmax()
min_hour =trips_per_hour.idxmin()
max_trip = trips_per_hour.max()
min_trip = trips_per_hour.min()
print(f"Maximum trip in the hour: {max_hour}")
print(f"Maximum trips: {max_trip}")
print(f"Minimun trip in the hour: {min_hour}")
print(f"Minimum trips: {min_trip}")
```

## Bar graph

```python
colors = ['blue']* len(trips_per_hour)
colors[list(trips_per_hour.index).index(max_hour)] = 'Red'
colors[list(trips_per_hour.index).index(min_hour)] = 'green'
plt.figure(figsize=(10,4))
# plt.subplot(1,1,1)
plt.bar(trips_per_hour.index,trips_per_hour.values, color =colors, edgecolor ="Black")
plt.text(max_hour,max_trip,str(max_trip), ha = "center", va ="bottom")
plt.text(min_hour,min_trip,str(min_trip), ha = "center", va ="bottom")
plt.xlabel("Hour of day")
plt.ylabel("No of trips")
plt.title("Trip distribution by hour")
plt.legend("Trip count")
plt.savefig("../../images/uber/hr_bar.png" )
plt.show()
```

## Line graph

```python
plt.figure(figsize=(10,5))
plt.scatter(max_hour, max_trip, color='black', s=50, label='Max Trips')
plt.scatter(min_hour, min_trip, color='blue', s=50, label='Min Trips')
plt.plot(trips_per_hour.index,trips_per_hour.values, color ="red", linewidth = 1, marker =".",label="Trips per hour")

plt.xlabel("Hour of day")
plt.ylabel("No of trips")
plt.title("Trip distribution by hour")
plt.legend()
plt.grid(True)
plt.savefig("../../images/uber/hr_line.png")
plt.show()
```

## OBSERVATION
#### The highest number of trips occurs at 7 PM, indicating peak demand in the evening, while the lowest number occurs at 5 AM, showing minimal activity early in the morning.

# WEEKDAY-WISE TRIP DISTRIBUTION

```python
df['day'] = df['pickup_datetime'].dt.day
df['weekday'] = df['pickup_datetime'].dt.day_name()

trips_per_weekday = df.groupby("weekday").size().reindex(weekday_order)
trips_per_weekday
```

```python

high_day = trips_per_weekday.idxmax()
high_day_trip = trips_per_weekday.max()

low_day = trips_per_weekday.idxmin()
low_day_trip = trips_per_weekday.min()

print(f"Highest trip in a week is {high_day_trip} on {high_day}")
print(f"Lowest trip in a week is {low_day_trip} on {low_day}")
```

## Bar graph

```python
colors = ["Green"]* len(trips_per_weekday)
colors[list(trips_per_weekday.index).index(high_day)] = 'Blue'
colors[list(trips_per_weekday.index).index(low_day)] = 'Red'
plt.figure(figsize=(10,5))
plt.bar(trips_per_weekday.index, trips_per_weekday.values, color = colors, label= "Trips")
plt.xlabel("WEEKDAY")
plt.ylabel("No of trips")
plt.title("Trip distribution by Weekday")


plt.text(high_day, high_day_trip, str(high_day_trip), va="bottom", ha ="center", label="lowest trip")
plt.text(low_day, low_day_trip, str(low_day_trip), va="bottom", ha ="center", label="highest trip")
plt.savefig("../../images/uber/day_bar.png")
plt.show()
```

## Line graph

```python
plt.figure(figsize=(10,5))
plt.plot(trips_per_weekday.index, trips_per_weekday.values, color = "red", marker = 'o', label= "Trips")
plt.scatter(high_day, high_day_trip, color = "black", s = 100, label="highest trip")
plt.scatter(low_day, low_day_trip, color = "green", s = 100, label="lowest trip")
plt.xlabel("weekday")
plt.ylabel("No of trips")
plt.title("Trips per weekday")
plt.legend()
plt.grid(True)
plt.savefig("../../images/uber/day_line.png")
plt.show()
```

## Trips by Weekday

#### The highest trips are on Friday (30,880 trips), suggesting end-of-week peak demand, possibly due to work travel, social events, or weekend preparations.

#### The lowest trips are on Monday (25,243 trips), indicating lighter travel at the start of the week.

#### The trend shows a gradual increase from Monday to Friday, slight decrease on Saturday, and further drop on Sunday.

## Implications:
#### Operational planning and surge pricing strategies can consider these weekday trends. For example, schedule more drivers towards the end of the workweek and fewer on Mondays.

# MONTH-WISE TRIP DISTRIBUTION

```python
monthwise_trip = df.groupby('month').size().sort_index()
monthwise_trip.index = monthwise_trip.index.map(lambda x : calendar.month_name[x])
monthwise_trip
```

```python
high_month = monthwise_trip.idxmax()
low_month = monthwise_trip.idxmin()

high_month_trip = monthwise_trip.max()
low_month_trip = monthwise_trip.min()

avg_trip = monthwise_trip.mean()
print(f"avg trip : {avg_trip}")

print(f"Highest trip is {high_month_trip} in the month {high_month}")
print(f"Lowest trip is {low_month_trip} in the month {low_month}")
```

## Bar graph

```python
plt.figure(figsize=(15,5))
colors=["Green"]*len(monthwise_trip)
colors[list(monthwise_trip.index).index(high_month)] = "Blue"
colors[list(monthwise_trip.index).index(low_month)] = "yellow"
plt.bar(monthwise_trip.index, monthwise_trip.values, color = colors, label="Trips")
plt.text(high_month, high_month_trip +0.5, str(high_month_trip), ha="center", va="bottom")
plt.text(low_month, low_month_trip +0.5, str(low_month_trip), ha="center", va="bottom")
plt.xlabel("Month")
plt.ylabel("No of trips")
plt.title("Month-wise trip distribution")
plt.legend()
plt.savefig("../../images/uber/month_bar.png")
plt.show()
```

## Line plot

```python
plt.figure(figsize=(10,5))
plt.plot(monthwise_trip.index, monthwise_trip.values, color = "red", marker = 'o', label= "Trips")
plt.scatter(high_month, high_month_trip, color = "black", s = 100, label="highest trip")
plt.scatter(low_month, low_month_trip, color = "green", s = 100, label="lowest trip")
plt.xlabel("Month")
plt.ylabel("No of trips")
plt.title("Trips per Month")
plt.legend()
plt.grid(True)
plt.savefig("../../images/uber/month_line.png")
plt.show()
```
